## Data Cleaning

In addition to individual countries, tehe `Country` column in this dataset also includes some levels that are not countries, but rather, are collections of countries grouped by some shared characteristic (typically geographical or economic in nature).

```python
data.Country.unique()    # List all unique levels
```

The code below excludes all "non-countries", restricting the analysis to individual countries only.

```python
excluded = ['Arab World', 'Caribbean small states', 'Central Europe and the Baltics', 'East Asia & Pacific (all income levels)',
'East Asia & Pacific (developing only)', 'Fragile and conflict affected situations', 'Heavily indebted poor countries (HIPC)',
'High income', 'High income: nonOECD', 'High income: OECD', 'Latin America & Caribbean (all income levels)', 
'Latin America & Caribbean (developing only)', 'Least developed countries: UN classification', 'Low & middle income',
'Low income', 'Lower middle income', 'Middle East & North Africa (all income levels)', 'Middle East & North Africa (developing only)',
'Middle income', 'OECD members', 'Other small states', 'Pacific island small states', 'Sub-Saharan Africa (all income levels)', 
'Sub-Saharan Africa (developing only)', 'World']
```

```python
data = data[~data.Country.isin(excluded)]
```

### Mobile Cellular Subscriptions

A simple data quality check can be constructed based on the number of mobile cellular subscriptions reported in the dataset.  According to Wikipedia, the world's first commercial cellular network was launched in Japan in 1979.  Next, commercial cellular service was launched by Nordic Mobile Telephone in Denmark, Finland, Norway, and Sweden in 1981.

The plot below shows the total number of cellular subscriptions per year in the world:

```python
# Group by year and sum cellular subscriptions
grp_by_cell = data.groupby(by='Year', as_index=False)[['Mobile Cellular Subscriptions']].sum()
# Only retain data up to 1996
grp_by_cell = grp_by_cell[grp_by_cell.Year <= 1996]
```

The function below helps control the frequency of axis tick labels shown.

```python
def show_nth_ticks(ax, n, axis):
    """Display only every nth axis tick label on the specified axis.
    ax: plot object
    n: only every nth axis tick label shown
    axis: can be either 'x' or 'y'
    """
    if axis == 'x':
        # first make every x axis tick label invisible
        [label.set_visible(False) for label in ax.xaxis.get_ticklabels()]
        # set every nth tick label to be visible
        [label.set_visible(True) for label in ax.xaxis.get_ticklabels()[::n]]
    elif axis == 'y':
        [label.set_visible(False) for label in ax.yaxis.get_ticklabels()]
        [label.set_visible(True) for label in ax.yaxis.get_ticklabels()[::n]]
```

```python
sns.set(style="ticks", rc={'figure.figsize':(16,9)}, font_scale=1.5)
ax = sns.barplot(x='Year', y='Mobile Cellular Subscriptions', data=grp_by_cell, color="salmon", saturation=.8)
ax.set(ylabel='Total Number of Cellular Subscriptions')
show_nth_ticks(ax, n=2, axis='x')
plt.title('Data quality check - mobile subscriptions up to 1996');
```

```python
# Group by year and sum cellular subscriptions per 100 People
grp_by_cell2 = data.groupby(by='Year', as_index=False)[['Mobile Cellular Subscriptions per 100 People']].sum()
# Only retain data up to 1996
grp_by_cell2 = grp_by_cell2[grp_by_cell2.Year <= 1996]
```

```python
sns.set(style="ticks", rc={'figure.figsize':(16,9)}, font_scale=1.5)
ax = sns.barplot(x='Year', y='Mobile Cellular Subscriptions per 100 People', data=grp_by_cell2, color="salmon", saturation=.5)
ax.set(ylabel='Total Number of Cellular Subscriptions per 100 People')
show_nth_ticks(ax, n=2, axis='x')
plt.title('Data quality check - mobile subscriptions per 100 people up to 1996');
```

The plots above show 2 dubious patterns:
* There are a large number of mobile cellular subscriptions prior to 1979 - which was the first year commercial cellular services were offered.
* Even focusing on only the data post-1979, the number of subscriptions shows the dubious pattern of first decreasing and then increasing.

Hence, it is reasonable to drop both columns related to mobile cellular subscriptions because the data in those columns cannot be trusted.

```python
data.drop(['Mobile Cellular Subscriptions', 'Mobile Cellular Subscriptions per 100 People'], axis=1, inplace=True)
```

### Telephone Lines

The plots in this section show a data quality check similar to the one in the previous section, applied to the data on the number of telephone lines.

```python
# Group by year and sum cellular subscriptions
grp_by_phone = data.groupby(by='Year', as_index=False)[['Telephone Lines']].sum()
```

```python
sns.set(style="ticks", rc={'figure.figsize':(16,9)}, font_scale=1.5)
ax = sns.barplot(x='Year', y='Telephone Lines', data=grp_by_phone, color="blue", saturation=.5)
ax.set(ylabel='Total Number of Telephone Lines')
show_nth_ticks(ax, n=5, axis='x')
plt.title('Data quality check - telephone lines');
```

```python
# Group by year and sum cellular subscriptions
grp_by_phone2 = data.groupby(by='Year', as_index=False)[['Telephone Lines per 100 People']].sum()
```

```python
sns.set(style="ticks", rc={'figure.figsize':(16,9)}, font_scale=1.5)
ax = sns.barplot(x='Year', y='Telephone Lines per 100 People', data=grp_by_phone2, color="blue", saturation=.5)
ax.set(ylabel='Total Number of Telephone Lines per 100 People')
show_nth_ticks(ax, n=5, axis='x')
plt.title('Data quality check - telephone lines per 100 people');
```

The plots above show some dubious patterns:
* Around 1980, there is first a descrease in the number of telephone lines, followed by an increase afterwards.
* There is a sudden unnaturally high jump in the number of telephone lines in 1980.  Also, the total number of telephone lines is shown to be more than 1 billion in 1980 itself - this is not supported by the data directly hosted on the World Bank's open data website, according to which the maximum number of telephone lines in the world has never even exceeded 1.5 billion (Source: https://data.worldbank.org/indicator/IT.MLT.MAIN)

Based on the above observations, it is safe to conclude that some type of data pollution occurred when the World Bank data was cleaned and hosted on the CORGIS Dataset Project.  It would be reasonable to drop the 2 features realted to telephones.



```python
data.drop(['Telephone Lines', 'Telephone Lines per 100 People'], axis=1, inplace=True)
```

## Data Exploration

### Distribution of Life Expectancy

Let's look at the distribution of the health indicator of a country's population - the life expectancy.  First, we look at the most recent data (2013):

```python
life_exp_2013 = data[data.Year == 2013]['Life Expectancy at Birth, Total']
```

```python
sns.set(color_codes=True, font_scale=1.5)
x = life_exp_2013.values
sns.distplot(x, bins=10, kde=False, rug=True, axlabel='Life Expectancy (Male & Female) at Birth (Years)')
plt.title('2013: Distribution of Life Expectancy (Male & Female) at Birth');
```

The histogram above shows the distribution of life expectancy of both males and females taken together for the year 2013.  A rugplot is also shown (small vertical dark blue tick marks right above the x axis). Every line in the rugplot represents an actual datapoint and thus gives an idea of the distribution of data - this can be valuable to know because the shape of the histogram can change based on the number of bins specified.

Here, it can be seen that in 2013, a significant number of countries in the world had their total life expectancy over 70 years.  Somewhat surprisingly, even in 2013, there were still 6 countries that had a life expectancy less than 50 years.

The plot below shows the variation of female life expectancy with male life expectancy.  Unsurprisingly, a linear trend emerges.

```python
sns.set(style="whitegrid", color_codes=True, font_scale=1.4)
df  = data[data.Year == 2013][['Life Expectancy at Birth, Female', 'Life Expectancy at Birth, Male']]
df.columns = ['Life Expectancy at Birth, Female (Years)', 'Life Expectancy at Birth, Male (Years)']

# scatter plot 
ax = sns.jointplot(x='Life Expectancy at Birth, Female (Years)', y='Life Expectancy at Birth, Male (Years)', data=df, 
                   size=8.5, stat_func=None, color='b', edgecolor='black', linewidth='1');
```

### Visualize Change in Life Expectancy with Other Features

'Arable land' is the land (in hectares per person in this dataset) that is actually suitable for agricultural activities.  On the other hand, 'agricultural land' is the land *legally designated* as being primarily used for agriculture.

The plot below shows the variation in the life expectancy with the arable land area in hectares per person of a country.

```python
sns.set(style="whitegrid", color_codes=True, font_scale=1.4)
df  = data[data.Year == 2013][['Arable Land', 'Life Expectancy at Birth, Total']]
df.columns = ['Arable Land (Hectares Per Person)', 'Life Expectancy at Birth, Total (Years)']

# scatter plot 
ax = sns.jointplot(x='Arable Land (Hectares Per Person)', y='Life Expectancy at Birth, Total (Years)', data=df, 
                   size=8.5, stat_func=None, color='g', edgecolor='black', linewidth='1');
```

In the plot above, there are 4 countries with very high values of arable land per person; these are distorting the plot and not allowing pattern in the rest of the data to emerge.  These 4 countries have been filtered out below and the plot above has been regenerated:

```python
sns.set(style="whitegrid", color_codes=True, font_scale=1.4)
df  = data[(data.Year == 2013) & (data['Arable Land'] <= 0.8)][['Arable Land', 'Life Expectancy at Birth, Total']]
df.columns = ['Arable Land (Hectares Per Person)', 'Life Expectancy at Birth, Total (Years)']

# scatter plot 
ax = sns.jointplot(x='Arable Land (Hectares Per Person)', y='Life Expectancy at Birth, Total (Years)', data=df, 
                   size=8.5, stat_func=None, color='g', edgecolor='black', linewidth='1');
```

We see that other than 2 outlier countries, a lot of countries with high life expectancy have medium-to-low arable land per person.  The two outlier countries are:

```python
data[(data.Year == 2013) & (data['Arable Land'] > 1.0)][['Country', 'Arable Land', 'Life Expectancy at Birth, Total']]
```

Since there is a difference between the definitions of arable land and agricultural land, the same plot shown above is regenerated, with the color encoding the value of agrcultural land:

```python
df  = data[(data.Year == 2013) & (data['Arable Land'] <= 0.8)][['Arable Land', 'Life Expectancy at Birth, Total', 'Arable Land Percent']]
df.columns = ['Arable Land (Hectares Per Person)', 'Life Expectancy at Birth, Total (Years)', 'Arable Land Percent']
x = df['Arable Land (Hectares Per Person)']
y = df['Life Expectancy at Birth, Total (Years)']
z = df['Arable Land Percent']

# set plot style
sns.set(style='darkgrid', rc={'figure.figsize':(10, 8.5)}, font_scale=1.5)

# for encoding a continuous varible as the color dimension, create a sequential palette
cmap = sns.cubehelix_palette(as_cmap=True)

f, ax = plt.subplots()

# assign the sequential color map to cmap to add the color dimension
points = ax.scatter(x, y, c=z, s=100, edgecolor='black', linewidth=1, cmap=cmap)
cbar = f.colorbar(points)      # insert a colorbar legend
cbar.set_label('Arable Land \n(% of \nLand Area)', rotation=0, labelpad=50)

# set axis labels
ax.set(ylabel='Life Expectancy at Birth, Total (Years)')
ax.set(xlabel='Arable Land (Hectares Per Person)');
```

No remarakable pattern emerges from the plot above.

```python
df = data[data.Year == 2013][['Life Expectancy at Birth, Male', 'Life Expectancy at Birth, Female', 'Life Expectancy at Birth, Total']]
df.columns = ['Males', 'Females', 'Combined']
# "Melt" the dataset to "long form" from the "wide form"
df = pd.melt(df, var_name='Life Expectancy', value_name='Years')
df.head()
```

```python
# Draw a categorical scatterplot to show each observation
sns.swarmplot(x="Life Expectancy", y="Years", color='navy', data=df)
plt.title('2013: Distribution of Life Expectancy at Birth for All Countries');
```

We see from the plot above that in 2013, a signifiacnt number of countries had hgih female life expectancy (75+ years).  The swarm plot also shows that there are more countries that fall in the higher range of life expectancy values than in the lower ranges.

### Change in the Distribution of Life Expectancy with Year 

The plot below displays how the life expectancy of males and females changes in its distribution across all countries in the world, for 3 different years.  Note that the years are spaced 2 decades apart each.

```python
df = data[data.Year.isin([1973, 1993, 2013])][['Year', 'Life Expectancy at Birth, Male', 'Life Expectancy at Birth, Female']]
df.columns = ['Year', 'Male', 'Female']
# "Melt" the dataset to "long form" from the "wide form"
df = pd.melt(df, id_vars='Year', var_name='Gender', value_name='Life Expectancy (Years)')
```

```python
sns.set(style="darkgrid", font_scale=1.4)
g = sns.FacetGrid(df, row="Gender", col="Year", size=4, margin_titles=True)
g.map(sns.distplot, "Life Expectancy (Years)", color="maroon", kde=True, bins=10)
g.set(xlim=(20, 100));
```

The facet grid plot above shows that life expectancy across the world for both males as well as females increased dramatically in the 2 decades between 1973 and 1993.  Furthermore, there was additional, though less dramatic, improvement in male and female life expectancy in the 2 decades between 1993 and 2013.

The information shown in the facet grid above can also be presented in the form of split violin plots, as shown below.

```python
sns.violinplot(x="Year", y="Life Expectancy (Years)", hue="Gender", data=df, split=True)
plt.title('Distribution of Life Expectancy for Males and Females for All Countries in 3 Different Years');
```

### Fertility Rate

The fertility rate of a country represents the average number of children a woman in that country would give birth to in her lifetime.

The plot below captures the variation of the fertility rate with the % of arable land in a country.  The data shown in this plot as well as the next one includes all years from 1960 - 2013; so each country is represented by several points (one per year).

The motivation for plotting the fertility rate against the arable land percentage is that in countries with a large amount of land suitable for farming (a labor intensive activity in many countries), there might potentially be a drive to have larger families.  However, the plot below shows that this is not true - countries with a large % arable land area have had fairly low fertility rates.

```python
df = data
y = df['Fertility Rate']
x = df['Arable Land Percent']

with sns.axes_style("ticks"):
    sns.jointplot(x=x, y=y, kind="kde", color="red", size=8, stat_func=None, space=0)
    
plt.title('Variation of Fertility Rate with % Arable Land Area: \nYears 1960 - 2013, All Countries', x = -2.5, y = 1.2);
```

Similar to the pot above, the plot below shows the variation of life expectancy with fertility rate.  The plot shows an almost-linear trend of total life expectancy at birth decreasing as the fertility rate increases.  This is possibly due to the fact that high fertility is often seen in economically under-developed countries.  Also, historically, as several countries developed and became prosperous, this development was often accompanied with a drop in fertility rates.

```python
df = data[['Fertility Rate', 'Life Expectancy at Birth, Total']]
df.columns = ['Fertility Rate', 'Life Expectancy at Birth, Total (Years)']
x = df['Fertility Rate']
y = df['Life Expectancy at Birth, Total (Years)']

with sns.axes_style("ticks"):
    g = sns.jointplot(x=x, y=y, kind="hex", color="coral", size=8, stat_func=None, space=0)

plt.title('Variation of Life Expectancy at Birth with Fertility Rate: \nYears 1960 - 2013, All Countries', x = -2.5, y = 1.2);
```
