# Data Cleaning

## Mobile Cellular Subscriptions

A simple data quality check can be constructed based on the number of mobile cellular subscriptions reported in the dataset.  According to Wikipedia, the world's first commercial cellular network was launched in Japan in 1979.  Next, commercial cellular service was launched by Nordic Mobile Telephone in Denmark, Finland, Norway, and Sweden in 1981.

The plot below shows the total number of cellular subscriptions per year in the world:

```python
# Group by year and sum cellular subscriptions
grp_by_cell = data.groupby(by='Year', as_index=False)[['Mobile Cellular Subscriptions']].sum()
# Only retain data up to 1996
grp_by_cell = grp_by_cell[grp_by_cell.Year <= 1996]
```

The function below helps control the frequency of axis tick labels shown.

```python
def show_nth_ticks(ax, n, axis):
    """Display only every nth axis tick label on the specified axis.
    ax: plot object
    n: only every nth axis tick label shown
    axis: can be either 'x' or 'y'
    """
    if axis == 'x':
        # first make every x axis tick label invisible
        [label.set_visible(False) for label in ax.xaxis.get_ticklabels()]
        # set every nth tick label to be visible
        [label.set_visible(True) for label in ax.xaxis.get_ticklabels()[::n]]
    elif axis == 'y':
        [label.set_visible(False) for label in ax.yaxis.get_ticklabels()]
        [label.set_visible(True) for label in ax.yaxis.get_ticklabels()[::n]]
```

```python
sns.set(style="ticks", rc={'figure.figsize':(16,9)}, font_scale=1.5)
ax = sns.barplot(x='Year', y='Mobile Cellular Subscriptions', data=grp_by_cell, color="salmon", saturation=.8)
ax.set(ylabel='Total Number of Cellular Subscriptions')
show_nth_ticks(ax, n=2, axis='x')
plt.title('Data quality check - mobile subscriptions up to 1996')
```

```python
# Group by year and sum cellular subscriptions per 100 People
grp_by_cell2 = data.groupby(by='Year', as_index=False)[['Mobile Cellular Subscriptions per 100 People']].sum()
# Only retain data up to 1996
grp_by_cell2 = grp_by_cell2[grp_by_cell2.Year <= 1996]
```

```python
sns.set(style="ticks", rc={'figure.figsize':(16,9)}, font_scale=1.5)
ax = sns.barplot(x='Year', y='Mobile Cellular Subscriptions per 100 People', data=grp_by_cell2, color="salmon", saturation=.5)
ax.set(ylabel='Total Number of Cellular Subscriptions per 100 People')
show_nth_ticks(ax, n=2, axis='x')
plt.title('Data quality check - mobile subscriptions per 100 people up to 1996')
```

The plots above show 2 dubious patterns:
* There are a large number of mobile cellular subscriptions prior to 1979 - which was the first year commercial cellular services were offered.
* Even focusing on only the data post-1979, the number of subscriptions shows the dubious pattern of first decreasing and then increasing.

Hence, it is reasonable to drop both columns related to mobile cellular subscriptions because the data in those columns cannot be trusted.

```python
data.drop(['Mobile Cellular Subscriptions', 'Mobile Cellular Subscriptions per 100 People'], axis=1, inplace=True)
```

## Telephone Lines

The plots in this section show a data quality check similar to the one in the previous section, applied to the data on the number of telephone lines.

```python
# Group by year and sum cellular subscriptions
grp_by_phone = data.groupby(by='Year', as_index=False)[['Telephone Lines']].sum()
```

```python
sns.set(style="ticks", rc={'figure.figsize':(16,9)}, font_scale=1.5)
ax = sns.barplot(x='Year', y='Telephone Lines', data=grp_by_phone, color="blue", saturation=.5)
ax.set(ylabel='Total Number of Telephone Lines')
show_nth_ticks(ax, n=5, axis='x')
plt.title('Data quality check - telephone lines')
```

```python
# Group by year and sum cellular subscriptions
grp_by_phone2 = data.groupby(by='Year', as_index=False)[['Telephone Lines per 100 People']].sum()
```

```python
sns.set(style="ticks", rc={'figure.figsize':(16,9)}, font_scale=1.5)
ax = sns.barplot(x='Year', y='Telephone Lines per 100 People', data=grp_by_phone2, color="blue", saturation=.5)
ax.set(ylabel='Total Number of Telephone Lines per 100 People')
show_nth_ticks(ax, n=5, axis='x')
plt.title('Data quality check - telephone lines per 100 people')
```

The plots above show some dubious patterns:
* Around 1980, there is first a descrease in the number of telephone lines, followed by an increase afterwards.
* There is a sudden unnaturally high jump in the number of telephone lines in 1980.  Also, the total number of telephone lines is shown to be more than 1 billion in 1980 itself - this is not supported by the data directly hosted on the World Bank's open data website, according to which the maximum number of telephone lines in the world has never even exceeded 1.5 billion (Source: https://data.worldbank.org/indicator/IT.MLT.MAIN)

Based on the above observations, it is safe to conclude that some type of data pollution occurred when the World Bank data was cleaned and hosted on the CORGIS Dataset Project.  It would be reasonable to drop the 2 features realted to telephones.



```python
data.drop(['Telephone Lines', 'Telephone Lines per 100 People'], axis=1, inplace=True)
```
