# Data Cleaning

A simple data quality check can be constructed based on the number of mobile cellular subscriptions reported in the dataset.  According to Wikipedia, the world's first commercial cellular network was launched in Japan in 1979.  Next, commercial cellular service was launched by Nordic Mobile Telephone in Denmark, Finland, Norway, and Sweden in 1981.

The plot below shows the total number of cellular subscriptions per year in the world:

```python
# Group by year and sum cellular subscriptions
grp_by_cell = data.groupby(by='Year', as_index=False)[['Mobile Cellular Subscriptions']].sum()
# Only retain data up to 1996
grp_by_cell = grp_by_cell[grp_by_cell.Year <= 1996]
```

```python
sns.set(style="ticks", rc={'figure.figsize':(15,9)}, font_scale=1.5)
ax = sns.barplot(x='Year', y='Mobile Cellular Subscriptions', data=grp_by_cell, color="salmon", saturation=.8)
ax.set(ylabel='Total Number of Cellular Subscriptions')
# show only alternate x axis tick labels
for label in ax.xaxis.get_ticklabels()[1::2]:
    label.set_visible(False)
plt.title('Data quality check - mobile subscriptions up to 1996')
```

```python
# Group by year and sum cellular subscriptions per 100 People
grp_by_cell2 = data.groupby(by='Year', as_index=False)[['Mobile Cellular Subscriptions per 100 People']].sum()
# Only retain data up to 1996
grp_by_cell2 = grp_by_cell2[grp_by_cell2.Year <= 1996]
```

```python
sns.set(style="ticks", rc={'figure.figsize':(15,9)}, font_scale=1.5)
ax = sns.barplot(x='Year', y='Mobile Cellular Subscriptions per 100 People', data=grp_by_cell2, color="salmon", saturation=.5)
ax.set(ylabel='Total Number of Cellular Subscriptions per 100 People')
# show only alternate x axis tick labels
for label in ax.xaxis.get_ticklabels()[1::2]:
    label.set_visible(False)
plt.title('Data quality check - mobile subscriptions up to 1996')
```

The plots above show 2 dubious patterns:
* There are a large number of mobile cellular subscriptions prior to 1979 - which was the first year commercial cellular services were offered.
* Even focusing on only the data post-1979, the number of subscriptions shows the dubious pattern of first decreasing and then increasing.

Hence, it is reasonable to drop both columns related to mobile cellular subscriptions because the data in those columns cannot be trusted.

```python
data.drop(['Mobile Cellular Subscriptions', 'Mobile Cellular Subscriptions per 100 People'], axis=1, inplace=True)
```
